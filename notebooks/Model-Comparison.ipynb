```python
metrics = ['Accuracy', 'Precision', 'Recall', 'F1']
logistic_regression = [0.84, 0.66, 0.20, 0.30]
naive_bayes = [0.66, 0.30, 0.70, 0.42]
decision_tree = [0.92, 0.78, 0.79, 0.78]
random_forest = [0.93, 0.85, 0.75, 0.80]
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.subplots(figsize = (12, 8))
bar_width = 0.2

bar1 = np.arange(len(metrics))
bar2 = [x + bar_width for x in bar1]
bar3 = [x + bar_width for x in bar2]
bar4 = [x + bar_width for x in bar3]

plt.bar(bar1, logistic_regression, width=bar_width, label="Logistic Regression")
plt.bar(bar2, naive_bayes, width=bar_width, label="Naive Bayes")
plt.bar(bar3, decision_tree, width=bar_width, label="Decision Tree")
plt.bar(bar4, random_forest, width=bar_width, label="Random Forest")

plt.yticks(np.arange(0.0, 1.0, 0.1))
plt.xticks([r + bar_width for r in range(len(metrics))], metrics)

plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.title("Model Comparison")
plt.xlabel("Evaulation Metric")
plt.ylabel("Score")
plt.legend(loc="best")
plt.show()
```
